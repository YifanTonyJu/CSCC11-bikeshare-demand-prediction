# Linear Regression Baseline

This notebook implements a baseline Linear Regression model for predicting hourly bike demand.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Load and Clean Data

We load the processed dataset and remove rows with missing values caused by lag features.

```python
df = pd.read_csv("../data/processed/final_dataset.csv")
df = df.dropna()

df.head()
```

## Train/Test Split

We use a time-based split:
- First 80% → training set  
- Last 20% → test set

```python
split_idx = int(len(df) * 0.8)

train = df.iloc[:split_idx]
test = df.iloc[split_idx:]
```

## Prepare Features and Train Model

We use all features except the target (`trips`) to train a Linear Regression model.

```python
X_train = train.drop(columns=["trips"])
y_train = train["trips"]

X_test = test.drop(columns=["trips"])
y_test = test["trips"]

model = LinearRegression()
model.fit(X_train, y_train)

print(model)
```

## Evaluation

We evaluate the model using MAE and RMSE.

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)
```

## Results

- MAE ≈ 3.83
- RMSE ≈ 6.88

## Analysis

The Linear Regression model captures general demand patterns,  
but its performance is limited due to its linear assumptions.

More advanced models such as Random Forest and XGBoost are expected to perform better.
